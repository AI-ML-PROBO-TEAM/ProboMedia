```python
text = """The extremely severe cyclonic storm “Biparjoy" has weakened and it is likely to cross between Mandvi (Gujarat) and Karachi (Pakistan) near Jakhau Port (Gujarat) by noon on 15th June as Very Severe Cyclonic Storm (VSCS). As a result, India Meteorological Department (IMD) forecasted heavy to very heavy falls over northeast India during next 5 days. An Orange alert has been issued for for Saurashtra and Kutch coasts as the cyclone is expected to make landfall in Gujarat on Thursday."""
```

```python
# Load tokenizer
# Importing dependencies from transformers
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
tokenizer = PegasusTokenizer.from_pretrained("google/pegasus-xsum")
# Load model
model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
```

```python
# Create tokens - number representation of our text
tokens = tokenizer(text, truncation=True, padding="longest", return_tensors="pt")
```

```python
# Input tokens
tokens
```

```python
# Summarize
summary = model.generate(**tokens)
```

```python
# Output summary tokens
summary[0]
```

```python
# Decode summary
tokenizer.decode(summary[0])
```
